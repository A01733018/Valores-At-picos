```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data= pd.read_csv('listings.csv') 

data.head(2)
```

```python
data1 = data.copy()
```

```python
#Sustituir valores nulos por un string en  concreto 
#description
data1["description"] =data1["description"].fillna("No description available")

#Neighborhood overview
data1["neighborhood_overview"] =data1["neighborhood_overview"].fillna("No description available")

#Host Location 
data1["host_location"] =data1["host_location"].fillna("Milan, Italy")

#host about 
data1["host_about"] =data1["host_about"].fillna("No description available")

#response time 
data1["host_response_time"] =data1["host_response_time"].fillna("N/A")

#response rate 
data1["host_response_rate"] =data1["host_response_rate"].fillna("N/A")

#response time 
data1["host_acceptance_rate"] =data1["host_acceptance_rate"].fillna("N/A")

#response time 
data1["host_is_superhost"] =data1["host_is_superhost"].fillna("f")

#host neighbourhood
data1["host_neighbourhood"] =data1["host_neighbourhood"].fillna("Not defined")

#neighbourhood
data1["neighbourhood"] =data1["neighbourhood"].fillna("Milan, Lombardy, Italy")


#cleansed
data1["neighbourhood_group_cleansed"] =data1["neighbourhood_group_cleansed"].fillna("Not defined")

#calendar updated
data1["calendar_updated"] =data1["calendar_updated"].fillna("No")

#availability  
data1["has_availability"] =data1["has_availability"].fillna("f")

#first review
data1["first_review"] =data1["first_review"].fillna("No date")

#last review
data1["last_review"] =data1["last_review"].fillna("No date")

#license
data1["license"] =data1["license"].fillna("Not included")
```

```python
#bathroom
data1["bathrooms"]= data1["bathrooms"].fillna(1) 
data1.head(5)

#bathroom
data1["bathrooms_text"]= data1["bathrooms_text"].fillna(1) 
data1.head(5)

#bedrooms 
data1["bedrooms"]= data1["bedrooms"].fillna(1) 
data1.head(5)


#beds
data1["beds"]= data1["beds"].fillna(1) 
data1.head(5)

#revenue 365 
data1["estimated_revenue_l365d"]= data1["estimated_revenue_l365d"].fillna(0) 
data1.head(5)

#review score 
data1["review_scores_rating"]= data1["review_scores_rating"].fillna(0) 
data1.head(5)

#review score accuracy 
data1["review_scores_accuracy"]= data1["review_scores_accuracy"].fillna(0) 
data1.head(5)

#review score cleanliness
data1["review_scores_cleanliness"]= data1["review_scores_cleanliness"].fillna(0) 
data1.head(5)

#review score check-inn
data1["review_scores_checkin"]= data1["review_scores_checkin"].fillna(0) 
data1.head(5)

#review score communication
data1["review_scores_communication"]= data1["review_scores_communication"].fillna(0) 
data1.head(5)

#review score location
data1["review_scores_location"]= data1["review_scores_location"].fillna(0) 
data1.head(5)

#review score location
data1["review_scores_value"]= data1["review_scores_value"].fillna(0) 
data1.head(5)

#review scores per month 
data1["reviews_per_month"]= data1["reviews_per_month"].fillna(0) 
data1.head(5)
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante forward fill" ("ffill")

data1["price"] =data1["price"].fillna(method="ffill") 
```

```python
#Identificar valores nulos por columna
valores_nulos=data1.isnull().sum()
pd.set_option("display.max_rows", None)       # Todas las filas
valores_nulos
```

```python
cuantitativas= data1.select_dtypes(include=["number"])
cualitativas = data1.select_dtypes(include=["object", "category"])
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
cuantitativas.plot(kind='box', vert=False)
plt.title("Valores AtÃ­picos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#MÃ©todo aplicando desviaciÃ³n estandar. Encuentro los valores extremos
y=cuantitativas
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data3= cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3.head(1)
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atÃ­picos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean.head(1)
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

**PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME**

```python
#MÃ©todo aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=cuantitativas

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr.head(1)
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atÃ­picos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.median(),1))
data4_iqr.head(1)
```

```python
# Unimos el dataframe cuantitativo limpio con el dataframe cualitativo

Datos_limpios = pd.concat([cualitativas, data4_iqr], axis=1)
Datos_limpios.head(1)
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Datos_limpios.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
Datos_limpios.to_csv("Act6.2.csv")
```
