```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
df= pd.read_csv('Ventas_totales.csv') 

df.head()
```

```python
df1=df.copy()
```

```python
#Corroboramos valores nulos
valores_nulos=df1.isnull().sum()
valores_nulos
```

Aplicar mÃ©todos para reemplazar valores nulos en cada columna

```python
df1["salon_ventas"]=df1["salon_ventas"].fillna(round(df["salon_ventas"].mean(),1))
df1["tarjetas_debito"]=df1["tarjetas_debito"].fillna(round(df["tarjetas_debito"].mean(),1))
df1["tarjetas_credito"]=df1["tarjetas_credito"].fillna(round(df["tarjetas_credito"].mean(),1))
df1["otros_medios"] =df1["otros_medios"].fillna(method="ffill") 
df1["subtotal_ventas_alimentos_bebidas"] =df1["subtotal_ventas_alimentos_bebidas"].fillna(method="ffill") 
df1["bebidas"] =df1["bebidas"].fillna(method="ffill") 
df1["almacen"] =df1["almacen"].fillna(method="ffill") 
df1["panaderia"] =df1["panaderia"].fillna(method="ffill") 
df1["lacteos"]=df1["lacteos"].fillna(round(df["lacteos"].median(),1))
df1["carnes"]=df1["carnes"].fillna(round(df["carnes"].median(),1))
df1["verduleria_fruteria"]=df1["verduleria_fruteria"].fillna(round(df["verduleria_fruteria"].median(),1))
df1["alimentos_preparados_rotiseria"]=df1["alimentos_preparados_rotiseria"].fillna(round(df["alimentos_preparados_rotiseria"].median(),1))
df1["indumentaria_calzado_textiles_hogar"]=df1["indumentaria_calzado_textiles_hogar"].fillna(round(df["indumentaria_calzado_textiles_hogar"].median(),1))
df1["electronicos_articulos_hogar"]=df1["electronicos_articulos_hogar"].fillna(round(df["electronicos_articulos_hogar"].median(),1))
df1["otros"]=df1["otros"].fillna(round(df["otros"].median(),1))
```

```python
#Creo 2 dataframe para poder procesar los outliers
cuantitativas= df1.iloc[ : , 1:23 ]
cualitativas= df1.iloc[ : , [0] ]
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
cuantitativas.plot(kind='box', vert=False)
plt.title("Valores AtÃ­picos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#MÃ©todo aplicando desviaciÃ³n estandar. Encuentro los valores extremos
y=cuantitativas
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
df2= cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df2
```

```python
#Reemplazamos valores atÃ­picos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=df2.copy()
data_clean=data_clean.fillna(round(df2.mean(),1))
data_clean.head()
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

**PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME**

```python
#MÃ©todo aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=cuantitativas

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del Dataframe
df3_iqr= cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
df3_iqr.head()
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=df3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atÃ­picos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
df4_iqr=df3_iqr.copy()
df4_iqr=df4_iqr.fillna(round(df3_iqr.median(),1))
df4_iqr.head()
```

```python
# Unimos el dataframe cuantitativo limpio con el dataframe cualitativo

Datos_limpios = pd.concat([cualitativas, df4_iqr], axis=1)
Datos_limpios.head()
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Datos_limpios.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
Datos_limpios.to_csv("ACT 6.1.csv")
```
